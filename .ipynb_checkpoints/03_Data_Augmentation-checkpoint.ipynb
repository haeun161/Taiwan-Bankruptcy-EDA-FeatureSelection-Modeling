## SMOTE를 활용한 데이터 증강
- 라벨 0인 데이터는 3419개로 다운샘플링
- 라벨 1인 데이터는 3400개로 오버샘플링 

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.utils import resample
```

```python
# 파일 경로 설정
file_path = "data.csv"

# CSV 파일을 DataFrame으로 불러오기
data = pd.read_csv(file_path)
```

```python
# Count the occurrences of each attack category
bankrupt_count = data['Bankrupt?'].value_counts()
print(bankrupt_count)
```

```python
# Majority class downsampling to 3419 instances
majority_data = data[data['Bankrupt?'] == 0]
downsampled_majority = resample(majority_data, replace=False, n_samples=3419, random_state=42)

# Combine the downsampled majority class with the original minority class
minority_data = data[data['Bankrupt?'] == 1]
combined_data = pd.concat([downsampled_majority, minority_data])

# Define features and target
X = combined_data.drop(columns=['Bankrupt?'])
y = combined_data['Bankrupt?']

# Minority class oversampling using SMOTE to reach 3400 instances
n_minority = len(minority_data)
sampling_strategy = {1: 3400}

smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Combine resampled data into a final DataFrame
resampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Bankrupt?'])], axis=1)

# Check the new class distribution
print(resampled_data['Bankrupt?'].value_counts())
```

```python
# Save the resampled data to a CSV file
resampled_data.to_csv('oversampled_data.csv', index=False)
```

```python
balanced_data = pd.read_csv('oversampled_data.csv')
print(balanced_data['Bankrupt?'].value_counts())
```

## SMOTE를 활용한 데이터 증강 2
- 라벨 0,1 6599개로 오버샘플링 

```python
# Count the occurrences of each attack category
bankrupt_count = data['Bankrupt?'].value_counts()
print(bankrupt_count)
```

```python
# Find the maximum count of attack categories
max_count = bankrupt_count.max()
print(max_count)
```

```python
# Instantiate SMOTE
smote = SMOTE(sampling_strategy={i: max_count for i in bankrupt_count.index})

# Oversample the dataset
X = data.drop(columns=['Bankrupt?'])
y = data['Bankrupt?']
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new balanced DataFrame
balanced_data = pd.concat([X_resampled, y_resampled], axis=1)

# Check the balanced count for each attack category
balanced_Bankrupt_counts = balanced_data['Bankrupt?'].value_counts()
print(balanced_Bankrupt_counts)

# Save the oversampled data to a CSV file
oversampled_file_path = 'oversampled_data1.csv'
balanced_data.to_csv(oversampled_file_path, index=False)

print(f'Oversampled data saved to {oversampled_file_path}')
```
